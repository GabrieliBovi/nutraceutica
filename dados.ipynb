```python
# Import
import pandas as pd  # data analysis

import matplotlib.pyplot as plt #data visualization
%matplotlib inline 

import seaborn as sns # data visualization

import numpy as np #numeric calculation

from IPython.display import display

from scipy.stats import linregress
```

```python
#abrindo arquivos - Grafico 1
df1 = pd.read_excel("Book2.xlsx", sheet_name = "Sheet1")
display(df1)
```

```python
#abrindo arquivos - Grafico 1
df2 = pd.read_excel("Book2.xlsx", sheet_name = "Sheet2")
display(df2)
```

```python
# Configurando o estilo do grfico
plt.style.use('ggplot')
plt.figure(figsize=(9, 6))
sns.set_style('white')

# Nomes
plt.ylabel('Papers', fontsize=14, fontfamily='serif')
plt.xlabel('Year (2008-2023)', fontsize=14, fontfamily='serif')

# Tamanho das letras dos eixos
plt.xticks(fontsize=11, fontfamily='serif')
plt.yticks(fontsize=11, fontfamily='serif')

# Plotando os dados de Flavonoids e Flavonoids and disease
plt.plot(df1.Year, df1.Flavonoids, label='Flavonoids',
         marker='o', lw='1', ms=5, fillstyle='full', color='salmon')
#plt.plot(df1.Year, df1.Disease, label='Flavonoids and disease',
         #marker='s', lw='1', ms=5, fillstyle='full', color='turquoise')

# Calculando a linha de regresso para Flavonoids de 2008 a 2018
slope_flavonoids_2008_2018, intercept_flavonoids_2008_2018, r_value_flavonoids_2008_2018, p_value_flavonoids_2008_2018, std_err_flavonoids_2008_2018 = linregress(df1.Year[df1.Year <= 2018], df1.Flavonoids[df1.Year <= 2018])

# Plotando a linha de regresso para Flavonoids de 2008 a 2018
plt.plot(df1.Year[df1.Year <= 2018], slope_flavonoids_2008_2018 * df1.Year[df1.Year <= 2018] + intercept_flavonoids_2008_2018, color='black', linestyle='--', lw=1)

# Calculando a linha de regresso para Flavonoids de 2018 a 2023
slope_flavonoids_2018_2023, intercept_flavonoids_2018_2023, r_value_flavonoids_2018_2023, p_value_flavonoids_2018_2023, std_err_flavonoids_2018_2023 = linregress(df1.Year[df1.Year >= 2018], df1.Flavonoids[df1.Year >= 2018])

# Plotando a linha de regresso para Flavonoids de 2018 a 2023
plt.plot(df1.Year[df1.Year >= 2018], slope_flavonoids_2018_2023 * df1.Year[df1.Year >= 2018] + intercept_flavonoids_2018_2023, color='black', linestyle='--', lw=1)


plt.legend(loc='upper left', fontsize=14, frameon=True, framealpha=0.5, facecolor='white')

# Centralizar os ticks do eixo x
plt.xticks(df1.Year)

#Gravar 
plt.savefig('flavonoids_9x6.tiff', dpi = 500)

# Exibindo o grfico
plt.show()

# Exibindo os resultados
print("Resultados para a linha de regresso de 2008 a 2018:")
print("Coeficiente angular:", slope_flavonoids_2008_2018)
print("Coeficiente linear:", intercept_flavonoids_2008_2018)
print("Coeficiente de determinao (R):", r_value_flavonoids_2008_2018**2)
print("\n")
print("Resultados para a linha de regresso de 2018 a 2023:")
print("Coeficiente angular:", slope_flavonoids_2018_2023)
print("Coeficiente linear:", intercept_flavonoids_2018_2023)
print("Coeficiente de determinao (R):", r_value_flavonoids_2018_2023**2)


```

```python
# Configurando o estilo do grfico
plt.style.use('ggplot')
plt.figure(figsize=(8, 5))
sns.set_style('white')

# Nomes
plt.ylabel('Papers', fontsize=14, fontfamily='serif')
plt.xlabel('Year (2008-2023)', fontsize=14, fontfamily='serif')

# Tamanho das letras dos eixos
plt.xticks(fontsize=12, fontfamily='serif')
plt.yticks(fontsize=12, fontfamily='serif')

# Plotando os dados de Flavonoids e Flavonoids and disease
plt.plot(df1.Year, df1.Flavonoids, label='Flavonoids',
         marker='o', lw='1', ms=5, fillstyle='full', color='salmon')
plt.plot(df1.Year, df1.Disease, label='Flavonoids and disease',
         marker='s', lw='1', ms=5, fillstyle='full', color='turquoise')

# Calculando a linha de regresso para Flavonoids de 2008 a 2018
slope_flavonoids_2008_2018, intercept_flavonoids_2008_2018 = np.polyfit(df1.Year[df1.Year <= 2018], df1.Flavonoids[df1.Year <= 2018], 1)

# Plotando a linha de regresso para Flavonoids de 2008 a 2018
plt.plot(df1.Year[df1.Year <= 2018], slope_flavonoids_2008_2018 * df1.Year[df1.Year <= 2018] + intercept_flavonoids_2008_2018, color='black', linestyle='--', label='Linear Regression (Flavonoids 2008-2018)', lw=1)

# Calculando a linha de regresso para Flavonoids de 2018 a 2023
slope_flavonoids_2018_2023, intercept_flavonoids_2018_2023 = np.polyfit(df1.Year[df1.Year >= 2018], df1.Flavonoids[df1.Year >= 2018], 1)

# Plotando a linha de regresso para Flavonoids de 2018 a 2023
plt.plot(df1.Year[df1.Year >= 2018], slope_flavonoids_2018_2023 * df1.Year[df1.Year >= 2018] + intercept_flavonoids_2018_2023, color='black', linestyle='--', label='Linear Regression (Flavonoids 2018-2023)', lw=1)

plt.legend(loc='upper left', fontsize=14, frameon=True, framealpha=0.5, facecolor='white')

# Centralizar os ticks do eixo x
plt.xticks(df1.Year)

# Exibindo o grfico
plt.show()
```
