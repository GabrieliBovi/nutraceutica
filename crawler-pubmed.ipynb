```python
import requests
```

```python
def buscar_informacoes_pubmed(filtro):
    base_url = 'https://pubmed.ncbi.nlm.nih.gov/?term='
    url = f'{base_url}{filtro}'

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        count_element = soup.find('span', class_='value')

        if count_element:
            count_str = count_element.text.strip()
            # Substitua vírgulas por pontos antes de tentar converter para inteiro
            count_str = count_str.replace(',', '.')
            count = int(float(count_str))  # Use float para lidar com decimais, se necessário
            return count
        else:
            print(f'Contagem não encontrada para o filtro "{filtro}". Retornando 0.')
            return 0
    else:
        print(f'Erro ao acessar o PubMed para o filtro "{filtro}". Código de status: {response.status_code}')
        return None
```

```python
def gerar_combinacoes_categorias(categorias):
    return list(combinations(categorias, 2))
```

```python
def criar_matriz_dados(categorias, anos):
    data = {'Categoria 1': [], 'Categoria 2': [], 'Ano': [], 'Contagem': []}

    for ano in range(min(anos), max(anos) + 1):
        for categoria1, categoria2 in gerar_combinacoes_categorias(categorias):
            filtro = f'{categoria1}+AND+{categoria2}+AND+("{ano}/01/01"[Date+-+Publication]:+"{ano}/12/31"[Date+-+Publication])'
            resultado = buscar_informacoes_pubmed(filtro)

            if resultado is not None:
                data['Categoria 1'].append(categoria1)
                data['Categoria 2'].append(categoria2)
                data['Ano'].append(ano)
                data['Contagem'].append(resultado)

    return pd.DataFrame(data)
```

```python
categorias = ['((Flavonoids [Title/Abstract] OR  Flavonoid[Title/Abstract]))' ,'((Nutraceutics[Title/Abstract])  OR  (Nutraceutical therapies[Title/Abstract]) OR (  Nutraceutical[Title/Abstract]  OR  (Nutraceutical therapy[Title/Abstract]))' ,'((Neurodegeneration[Title/Abstract]  OR  Neurodegenerative[Title/Abstract]))' ,'((Disease[Title/Abstract]  OR  Diseases[Title/Abstract]))' ,'((Aging[Title/Abstract]))' ,'((Anti aging[Title/Abstract]))' ,'((Parkinson’s disease[Title/Abstract]  OR  PD[Title/Abstract]))' ,'((Alzheimer’s disease[Title/Abstract]  OR  AD[Title/Abstract]))' ,'((Amyotrophic lateral sclerosis[Title/Abstract]  OR  ALS[Title/Abstract]))' ,'((Huntington’s disease[Title/Abstract] OR  HD[Title/Abstract]))' ,'((Dementia[Title/Abstract]))' ,'((Anti-inflammatory [Title/Abstract]))' ,'((Antioxidant[Title/Abstract]))' ,'((Oxidative stress[Title/Abstract]  OR  ROS [Title/Abstract]) OR (Reactive oxygen species[Title/Abstract]))' ,'((Retinitis pigmentosa[Title/Abstract]  OR  RP[Title/Abstract]))' ,'((Retina[Title/Abstract]))' ,'((Age-related macular degeneration[Title/Abstract] OR  AMD[Title/Abstract]))' ,'((Neuroprotective[Title/Abstract]))' ,'((Neuroinflammation[Title/Abstract]))' ,'((Photoreceptor[Title/Abstract]))' ,'((Polyphenols[Title/Abstract]  OR  Polyphenol[Title/Abstract]))']
anos = [2003, 2023]
```

```python
matriz_dados = criar_matriz_dados(categorias, anos)
```
